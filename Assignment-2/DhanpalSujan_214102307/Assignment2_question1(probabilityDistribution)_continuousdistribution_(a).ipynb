```python
import numpy
import matplotlib.pyplot as plt
from numpy import random
```

```python
x=numpy.arange(-10,10,0.1)
mu1=2
sigma1=3
y1=[]

mu2=5
sigma2=2
y2=[]

mu3=0
sigma3=1
y3=[]

for i in x:
    fun=1/(sigma1*numpy.sqrt(2*numpy.pi))*numpy.exp(-0.5*(i-mu1)**2/(sigma1**2))
    y1.append(fun)
    
for i in x:
    fun=1/(sigma2*numpy.sqrt(2*numpy.pi))*numpy.exp(-0.5*(i-mu2)**2/(sigma2**2))
    y2.append(fun)

for i in x:
    fun=1/(sigma3*numpy.sqrt(2*numpy.pi))*numpy.exp(-0.5*(i-mu3)**2/(sigma3**2))
    y3.append(fun)



    
y1=numpy.array(y1)
y2=numpy.array(y2)
y3=numpy.array(y3)

plt.plot(x,y1,color='red')    
plt.plot(x,y2,color='blue')
plt.plot(x,y3,color='green')
plt.plot("Gaussian distribution pdf plot")
plt.xlabel("x value")
plt.ylabel(" PDF values")

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,sharey=True, figsize=(15,4))

ax1.plot(x, y1, 'tab:red')
ax1.set_xlim(-10,10)
ax1.axvline(x=mu1,color='black')
ax1.axhline(y=1.0*numpy.exp(-0.5)/numpy.sqrt(2*numpy.pi*sigma1*sigma1),xmin=(mu1+10)/20,xmax=(mu1+sigma1+10)/20,color='brown')


ax2.plot(x,y2, 'tab:blue')
ax2.set_xlim(-10,10)
ax2.axvline(x=mu2,color='black')
ax2.axhline(y=1.0*numpy.exp(-0.5)/numpy.sqrt(2*numpy.pi*sigma2*sigma2),xmin=(mu2+10)/20,xmax=(mu2+sigma2+10)/20,color='brown')


ax3.plot(x, y3, 'tab:green')
ax3.axvline(x=mu3,color='black',label='mean')
ax3.axvline(x=mu3,color='black',label='mode')
ax3.axvline(x=mu3,color='black',label='median')
ax3.set_xlim(-10,10)
ax3.axhline(y=1.0*numpy.exp(-0.5)/numpy.sqrt(2*numpy.pi*sigma3*sigma3),xmin=(mu3+10)/20,xmax=(mu3+sigma3+10)/20,color='brown',label='std dev')
plt.legend(fontsize=10)

for ax in fig.get_axes():
    ax.label_outer()
```

```python
x=numpy.arange(0,10,0.1)
lambda1=2
E1=[]

lambda2=1.5
E2=[]

lambda3=0.5
E3=[]

def mean(L):
    return 1/L
    
def median(L):
    return numpy.log(2)/L
def variance(L):
    return 1/L**2

for i in x:
    fun= lambda1*numpy.exp(-lambda1*i)
    E1.append(fun)

for i in x:
    fun= lambda2*numpy.exp(-lambda2*i)
    E2.append(fun)
    
for i in x:
    fun= lambda3*numpy.exp(-lambda3*i)
    E3.append(fun)
    

    
E1=numpy.array(E1)
E2=numpy.array(E2)
E3=numpy.array(E3)

plt.plot(x,E1,color='red')
plt.plot(x,E2, color='blue')
plt.plot(x,E3, color='green')
plt.plot("Exponential distribution pdf plot")
plt.xlabel("x value")
plt.ylabel(" PDF values")

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,sharey=True, figsize=(15,4))

ax1.plot(x, E1, 'tab:red')
ax1.axvline(x=mean(lambda1), color='black')
ax1.axvline(x=median(lambda1), color='brown', linestyle='dashed')
ax1.set_xlim(0,10)
ax1.axhline(y=lambda1/numpy.exp(1), xmin=0, xmax=((1/lambda1))/10, color='orange')

ax2.plot(x,E2, 'tab:blue')
ax2.axvline(x=mean(lambda2), color='black')
ax2.axvline(x=median(lambda2), color='brown', linestyle='dashed')
ax2.set_xlim(0,10)
ax2.axhline(y=lambda2/numpy.exp(1), xmin=0, xmax=((1/lambda2))/10, color='orange')



ax3.plot(x, E3, 'tab:green')
ax3.axvline(x=mean(lambda3), color='black',label='mean')
ax3.axvline(x=median(lambda3), color='brown', linestyle='dashed',label='mdian')
ax3.set_xlim(0,10)
ax3.axhline(y=lambda3/numpy.exp(1), xmin=0, xmax=((1/lambda3))/10, color='orange',label='std dev')
plt.legend()

for ax in fig.get_axes():
    ax.label_outer()
```

```python
x=numpy.arange(0,10,0.1)
alpha1=1
beta1=3
B1=[]

alpha2=1
beta2=1
B2=[]

alpha3=2
beta3=2
B3=[]

def mean(a,b):
    return a/(a+b)


        
for i in x:
    fun=(i**(alpha1-1)*(1-i)**(beta1-1))*(numpy.math.factorial(alpha1+beta1-1))/((numpy.math.factorial(alpha1-1))*(numpy.math.factorial(beta1-1)))
    B1.append(fun)
    
for i in x:
    fun=(i**(alpha2-1)*(1-i)**(beta2-1))*(numpy.math.factorial(alpha2+beta2-1))/((numpy.math.factorial(alpha2-1))*(numpy.math.factorial(beta2-1)))
    B2.append(fun)
    
for i in x:
    fun=(i**(alpha3-1)*(1-i)**(beta3-1))*(numpy.math.factorial(alpha3+beta3-1))/((numpy.math.factorial(alpha3-1))*(numpy.math.factorial(beta3-1)))
    B3.append(fun)
    
B1=numpy.array(B1)
B2=numpy.array(B2)
B3=numpy.array(B3)

plt.plot(x,B1,color='red')
plt.plot(x,B2,color='blue')
plt.plot(x,B3,color='green')
plt.plot("Beta distribution pdf plot")
plt.xlabel("x value")
plt.ylabel(" PDF values")

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,sharey=True, figsize=(15,4))
#fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, B1, 'tab:red')
ax1.axvline(x=mean(alpha1,beta1), color='black')

ax2.plot(x,B2, 'tab:blue')
ax2.axvline(x=mean(alpha2,beta2), color='black')

ax3.plot(x, B3, 'tab:green')
ax3.axvline(x=mean(alpha3,beta3), color='black',label='mean')
plt.legend()
for ax in fig.get_axes():
    ax.label_outer()
```

```python
x=numpy.arange(0,10,0.1)
k1=2
C1=[]

k2=4
C2=[]

k3=8
C3=[]

for i in x:
    fun=(i**(0.5*k1-1)*numpy.exp(-0.5*i))/((2**(0.5*k1))*(numpy.math.factorial(0.5*k1-1)))
    C1.append(fun)
    
for i in x:
    fun=(i**(0.5*k2-1)*numpy.exp(-0.5*i))/((2**(0.5*k2))*(numpy.math.factorial(0.5*k2-1)))
    C2.append(fun)
    
for i in x:
    fun=(i**(0.5*k3-1)*numpy.exp(-0.5*i))/((2**(0.5*k3))*(numpy.math.factorial(0.5*k3-1)))
    C3.append(fun)

def median(k):
    return k*(1-2/(9*k))**3

def mode(k):
    return max((k-2),0)

def variance(k):
    return 2*k

C1=numpy.array(C1)
C2=numpy.array(C2)
C3=numpy.array(C3)

plt.plot(x,C1,color='red')

plt.plot(x,C2,color='blue')
plt.plot(x,C3,color='green')
plt.title("Chi squarred pdf plot")
plt.xlabel("x value")
plt.ylabel(" PDF values")

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,sharey=True, figsize=(15,4))
ax1.plot(x, C1, 'tab:red')
ax1.axvline(x=k1, color='black')
ax1.axvline(x=median(k1), color='brown',linestyle='dashed')
ax1.axvline(x=mode(k1), color='yellow', linestyle='dotted')
ax1.set_xlim(0,10)
ax1.axhline(y=numpy.sqrt(2*k1)**(0.5*k1-1)*numpy.exp(-numpy.sqrt(k1/2))/((2**(0.5*k1))*(numpy.math.factorial(0.5*k1-1))), xmin=0, xmax=numpy.sqrt(2*k1)/10, color='orange')


ax2.plot(x,C2, 'tab:blue')
ax2.axvline(x=k2, color='black')
ax2.axvline(x=median(k2), color='brown',linestyle='dashed')
ax2.axvline(x=mode(k2), color='yellow', linestyle='dotted')
ax2.set_xlim(0,10)
ax2.axhline(y=numpy.sqrt(2*k2)**(0.5*k2-1)*numpy.exp(-numpy.sqrt(k2/2))/((2**(0.5*k2))*(numpy.math.factorial(0.5*k2-1))), xmin=0, xmax=numpy.sqrt(2*k2)/10, color='orange')


ax3.plot(x, C3, 'tab:green')
ax3.axvline(x=k2, color='black',label='mean')
ax3.axvline(x=median(k3), color='brown',linestyle='dashed',label='median')
ax3.axvline(x=mode(k3), color='yellow', linestyle='dotted',label='mode')
ax3.set_xlim(0,10)
ax3.axhline(y=numpy.sqrt(2*k3)**(0.5*k3-1)*numpy.exp(-numpy.sqrt(k3/2))/((2**(0.5*k3))*(numpy.math.factorial(0.5*k3-1))), xmin=0, xmax=numpy.sqrt(2*k3)/10, color='orange',label='std dev')
plt.legend()

for ax in fig.get_axes():
    ax.label_outer()
```

```python
x= numpy.arange(-10,10,0.1)
mu=0
sigma=1
N=[]

for i in x:
    fun=1/(sigma*numpy.sqrt(2*numpy.pi))*numpy.exp(-0.5*(i-mu)**2/(sigma**2))
    N.append(fun)

N=numpy.array(N)
plt.plot(x,N,color='red')
plt.axvline(x=mu, color='black',label='mean,median,mode')
plt.title("standard normal pdf plot")
plt.xlabel("x value")
plt.ylabel(" PDF values")
plt.axhline(y=1.0*numpy.exp(-0.5)/numpy.sqrt(2*numpy.pi*sigma*sigma),xmin=(mu+10)/20,xmax=(mu+sigma+10)/20,color='brown',label='std dev')
plt.legend()
```

```python
a1=2
b1=4
U1=[]

a2=5
b2=9
U2=[]

a3=3
b3=4
U3=[]

def mean(a,b):
    return (a+b)/2

def median(a,b):
    return (a+b)/2

def var(a,b):
    return (b-a)**2/12

def uni(a,b,x):
    if ((a<=x) and (x<=b)):
        return 1.0/(b-a)
    else:
        return 0

x=numpy.arange(0,10,0.1)
for i in x:
    U1.append(uni(a1,b1,i))
    U2.append(uni(a2,b2,i))
    U3.append(uni(a3,b3,i))

U1=numpy.array(U1)
U2=numpy.array(U2)
U3=numpy.array(U3)

plt.plot(x,U1,color='red')
plt.plot(x,U2,color='blue')
plt.plot(x,U3,color='green')
plt.title("Uniform distribution pdf plot")
plt.xlabel("x value")
plt.ylabel(" PDF values")

fig, (ax1, ax2,ax3) = plt.subplots(1, 3,sharey=True, figsize=(15,4))
ax1.plot(x, U1, 'tab:red')
ax1.axvline(x=mean(a1,b1),color='black')
ax1.axvline(x=median(a1,b1),color='brown',linestyle='dashed')


ax2.plot(x,U2, 'tab:blue')
ax2.axvline(x=mean(a2,b2),color='black')
ax2.axvline(x=median(a2,b2),color='brown',linestyle='dashed')



ax3.plot(x,U3, 'tab:green')
ax3.axvline(x=mean(a3,b3),color='black', label='mean')
ax3.axvline(x=median(a3,b3),color='brown',linestyle='dashed')
plt.legend()
```
