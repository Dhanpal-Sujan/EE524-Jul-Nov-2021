```python
import numpy
import matplotlib.pyplot as plt
from numpy import random
```

```python
Hassian_mat=numpy.array([[2,1],[1,2]]) #given hessian matrix
eigval,eigvec=numpy.linalg.eig(Hassian_mat) #calculate and stores the eigen value and eigen vector 
print("eigen values are",eigval)
print("The quadratic form extracted from hessian matrix is:") #the possible quadratic form
print("x1**2+x2**2+x1x2+ax1+bx2+c")
print("the covexcity is strictly convex") # the eigen values of the matrix are greater than 0. so the matrix is positive definite. 
#so the convexity of function is strictly convex

a=2
b=3
c=5
```

#### The eigen values of the matrix are greater than 0. so the matrix is positive definite. So, the convexity of function is strictly convex

#### The possible quadratic form is x1^2+x2^2+x1x2+ax1+bx2+c

```python
def fun(x1,x2):
    return x1**2+x2**2+x1*x2+a*x1+b*x2+c
```

```python
x1=numpy.arange(-5,5,0.1)
x2=numpy.arange(-5,5,0.1)
x1temp=[]
x2temp=[]
y=[]
for i in x1:
    for j in x2:
        x1temp.append(i)
        x2temp.append(j)
        y.append(fun(i,j))

x1temp=numpy.array(x1temp)
x2temp=numpy.array(x2temp)
y=numpy.array(y)
```

```python
ax=plt.axes(projection='3d')
ax.plot3D(x1temp,x2temp,y,'o')
ax.plot(x1temp,x2temp,y,'o')

plt.title("plot of quadratic form")
plt.xlabel("x1")
plt.ylabel("x2")
```
