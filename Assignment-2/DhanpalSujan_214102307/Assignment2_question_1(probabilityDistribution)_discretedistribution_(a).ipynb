```python
import numpy
import matplotlib.pyplot as plt
from math import comb
```

```python
pf1=0.1
ps1=0.9

x1=numpy.arange(0,2,1)
y1=[]

s1=0
s11=0
for i in x1:
    fun=comb(1,i)*0.9**i*0.1**(1-i)
    y1.append(fun)
for i in x1:
    s1=s1+i*comb(1,i)*0.9**i*0.1**(1-i)
for i in x1:
    s11=s11+i*i*comb(1,i)*0.9**i*0.1**(1-i)
    
var=s11-s1*s1
print("The variance of random variable X is: ", var)

print("The mean of random variable X is",s1)
print(y1)
plt.plot(x1,y1)
plt.title("plot of PMF")
```
