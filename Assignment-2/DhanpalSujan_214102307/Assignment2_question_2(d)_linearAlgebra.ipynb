```python
def f(x1,x2,mat): # this function calculates x(transpose)Ax
    x = numpy.array([x1,x2])
    x = x.reshape((2,1))
    return numpy.dot(numpy.dot(x.T,mat),x)
```

```python
x1=numpy.arange(-5,5,0.1)
x2=numpy.arange(-5,5,0.1)
x11temp=[]
x21temp=[]
y1=[]
for i in x1:
    for j in x2:
        x11temp.append(i)
        x21temp.append(j)
        y1.append(f(i,j,H1)[0][0])

y1=numpy.array(y1)
x11temp=numpy.array(x11temp)
x21temp=numpy.array(x21temp)


print(x11temp[numpy.argmin(y1)])
print(x21temp[numpy.argmin(y1)])
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(x11temp,x21temp,y1, 'o')
#ax.view_init(0, 90)
ax.plot(x11temp,x21temp,y1,'o')
plt.title("plot of quadratic equation of 1st matrix")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
