```python
import numpy
import matplotlib.pyplot as plt
from numpy import random
from math import comb
```

```python
pf=0.4 # prob of receiving 0
ps=0.6 # prob of receiving 1
x=numpy.arange(0,21,1) #It is receiving 20 bits
y=[]

# The distribution it follows is binomial distribution

s=0
s1=0
for i in x:
    fun=comb(20,i)*(0.6)**i*0.4**(20-i)
    y.append(fun)
    
for i in x:
    s=s+i*comb(20,i)*0.6**i*0.4**(20-i)
for i in x:
    s1=s1+i*i*comb(20,i)*0.6**i*0.4**(20-i)
var=s1-s*s
print("The variance of random variable X is:",var)
print("the mean of random variable X is:",s)
y=numpy.array(y)

plt.plot(x,y,marker='o')
plt.title("plot of PMF")
plt.xlabel("values of 20 bits ")
plt.ylabel("values of corresponding PMF's")
```
