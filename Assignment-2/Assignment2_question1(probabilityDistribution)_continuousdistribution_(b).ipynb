```python
import numpy
import matplotlib.pyplot as plt
from numpy import random
```

```python
mu, sigma = 0, 2 # the mean is 0 and standard deviation is 2
x = numpy.random.normal(mu, sigma, 1000000) #It draws the samples from normal distribution
y=numpy.random.normal(mu, sigma, 1000000) #It draws the samples from normal distribution
```

```python
count, bins, ignored = plt.hist(x, 50, density=True)
plt.plot(bins, 1/(sigma * numpy.sqrt(2 * numpy.pi)) *
               numpy.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
t=x**2+y**2
count, bins, ignored = plt.hist(t, 50, density=True)
k1=2
i=bins/4
plt.plot(bins, ((i**(0.5*k1-1)*numpy.exp(-0.5*i))/((2**(0.5*k1))*(numpy.math.factorial(0.5*k1-1))))/4,linewidth=2, color='r')
plt.title("plot of X**2 +Y**2")

#The sum of squares of two orthogonal guassian random variable follow chi-square distribution

plt.show()
```
