```python
import numpy
import matplotlib.pyplot as plt
import pandas
from  numpy import random
```

```python
a=2
b=3
sigma=4
y=[]
x=numpy.arange(-10,10,0.2)
for i in x:
    fun=a*i+b
    y.append(fun)

N=[]
y1=[]

def normal(p):
    return numpy.exp(-0.5*p*p)/numpy.sqrt(2*numpy.pi)
for i in x:
    N.append(normal(i))

for i in x:
    fun1=a*i+b+sigma*numpy.random.randn() 
    y1.append(fun1)


y1=numpy.array(y1)
N=numpy.array(N)
y=numpy.array(y)

print(len(N))
plt.plot(x,y,color='black')
plt.plot(x,y1,'o',color='red')
plt.title("line y=2x+3")
plt.xlabel("x values ")
plt.ylabel("corresponding y values")
```
