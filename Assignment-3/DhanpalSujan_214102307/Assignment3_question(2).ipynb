```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy import random
```

```python
#x=np.arange(-1,1,0.01)
#y = 2*x+3+np.random.randn(x.shape[0])

x1=np.loadtxt("test.out",dtype='float',delimiter=None)
y1=np.loadtxt("test.out1",dtype='float',delimiter=None)
```

```python

a=numpy.arange(-10,10,0.1)
b=numpy.arange(-10,10,0.1)
atemp=[]
btemp=[]
E=[]

for i in a:
    for j in b:
        atemp.append(i)
        btemp.append(j)
        k=np.mean((y1-(i*x1+j))**2)
        E.append(k)


E=numpy.array(E)
atemp=numpy.array(atemp)
btemp=numpy.array(btemp)
E2d=E.reshape(200,200)
```

```python
X,Y=numpy.meshgrid(a,b)
fig, ax = plt.subplots(1, 1)
ax.contour(X, Y, E2d,levels=[1,2,4,8,16,32,64,128,256])
ax.set_xlim(-10,10)
ax.set_ylim(-1,5)
plt.show()
```
